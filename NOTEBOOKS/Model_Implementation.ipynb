Loading Dataframes

```python
#note that hi1 and lo1 will denote models for DecisionTreeClassifier
#note that hi2 and lo2 will denote models for MLPClassifier
#note that hi3 and lo3 will denote models for HistGradientBoostingClassifier

import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, clear_output
from joblib import Parallel, delayed
import joblib


MLP_Highest = {'solver': 'adam', 'max_iter': 1000, 'learning_rate_init': 0.001, 'learning_rate': 'adaptive', 'hidden_layer_sizes': (50,), 'alpha': 0.01, 'activation': 'relu'}
MLP_Lowest = {'solver': 'adam', 'max_iter': 1000, 'learning_rate_init': 0.001, 'learning_rate': 'invscaling', 'hidden_layer_sizes': (100,), 'alpha': 0.001, 'activation': 'relu'}

Highest_Best_parameters= {'criterion': 'gini', 'max_depth': 10, 'min_samples_leaf': 2, 'min_samples_split': 5}
Lowest_Best_parameters= {'criterion': 'gini', 'max_depth': None, 'min_samples_leaf': 1, 'min_samples_split': 2}

lo_hist_param =  {'min_samples_leaf': 1, 'max_leaf_nodes': 60, 'max_iter': 1000, 'max_depth': None, 'max_bins': 30, 'learning_rate': 0.01, 'l2_regularization': 0.01}
hi_hist_param =  {'min_samples_leaf': 1, 'max_leaf_nodes': 60, 'max_iter': 1000, 'max_depth': None, 'max_bins': 60, 'learning_rate': 0.01, 'l2_regularization': 0}

```

Loading Training Sets

```python
X1_train = pd.read_csv(data_dir + 'X1_train.csv')
X2_train = pd.read_csv(data_dir + 'X2_train.csv')
lo1_train = pd.read_csv(data_dir + 'lo1_train.csv')
lo2_train = pd.read_csv(data_dir + 'lo2_train.csv')
hi1_train = pd.read_csv(data_dir + 'hi1_train.csv')
hi2_train = pd.read_csv(data_dir + 'hi2_train.csv')
```

Decision Tree Classifier Models

Highest Region Game Sales

```python
from sklearn.tree import DecisionTreeClassifier
hdt = DecisionTreeClassifier(**Highest_Best_parameters, random_state=427)
hdt = hdt.fit(X1_train, hi1_train)
joblib.dump(hdt, model_dir +'Highest_Region_DecisionTreeClassifier.pkl')
```

Lowest Region Game Sales

```python
ldt = DecisionTreeClassifier(**Highest_Best_parameters, random_state=427)
ldt = ldt.fit(X1_train, lo1_train)
joblib.dump(ldt, model_dir +'Lowest_Region_DecisionTreeClassifier.pkl')
```

MLP Classifier Models

Highest Region Game Sales

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.neural_network import MLPClassifier
scaler = MinMaxScaler(feature_range = (0,1))
X1_train = scaler.fit_transform(X1_train)
hmlp = MLPClassifier(**MLP_Highest, random_state=427)
hmlp = hmlp.fit(X1_train, hi1_train)
joblib.dump(hmlp, model_dir +'Highest_Region_MLPClassifier.pkl')
```

Lowest Region Game Sales

```python
lmlp = MLPClassifier(**MLP_Lowest, random_state=427)
lmlp = lmlp.fit(X1_train, lo1_train)
joblib.dump(lmlp, model_dir +'Lowest_Region_MLPClassifier.pkl')
```

Hist Gradient Boosting Classifier

Highest Region Game Sales

```python
from sklearn.ensemble import HistGradientBoostingClassifier

hhgb = HistGradientBoostingClassifier(**hi_hist_param, random_state=427)
hhgb = hhgb.fit(X2_train, hi2_train)
joblib.dump(hhgb, model_dir +'Highest_Region_HistGradientBoostingClassifierClassifier.pkl')
```

Lowest Region Game Sales

```python
lhgb = HistGradientBoostingClassifier(**lo_hist_param, random_state=427)
lhgb = hhgb.fit(X2_train, lo2_train)
joblib.dump(lhgb, model_dir +'Lowest_Region_HistGradientBoostingClassifierClassifier.pkl')
```
