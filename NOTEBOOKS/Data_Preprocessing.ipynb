```python
has_null = [row.isnull().any() for _, row in df.iterrows()]
no_null = [not c for c in has_null]
print(df[has_null].describe())
print(df[no_null].describe())
```

Create new columns for Highest and Lowest Regions

```python
region = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
def find_highest_lowest(row):
  low = None
  high = None
  for reg in region:
    if low == None:
      low = (row[reg], reg)
    if low[0] > row[reg]:
      low = (row[reg], reg)
    if high == None:
      high = (row[reg], reg)
    if high[0] < row[reg]:
      high = (row[reg], reg)

  return pd.Series({'highest': high[1], 'lowest': low[1]})

# Apply the function row-wise to the DataFrame and store the results in new columns
df[['highest', 'lowest']] = df.apply(find_highest_lowest, axis=1)
df = df.drop(['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Name'], axis = 1)
```

One-Hot Encoding of Categorical Values

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize OneHotEncoder
encoder = OneHotEncoder()

categories = ['Platform',	'Year_of_Release',	'Genre',	'Publisher',	'Developer',	'Rating']
targets = ['highest', 'lowest']

# Perform one-hot encoding
encoded_array = encoder.fit_transform(df[categories]).toarray()

# Create a DataFrame from the encoded array
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categories))

# Concatenate the original DataFrame with the encoded DataFrame
# df2 for HistGradientBoostingClassifier, has null values
df2 = pd.concat([df, encoded_df], axis=1)

# Convert string values to NaN
df2['User_Score'] = pd.to_numeric(df2['User_Score'], errors='coerce')

df2 = df2.drop(categories, axis = 1)

# df1 for MLPClassifier, has no null values
df1 = df2[no_null]

print("Original DataFrame:")
print(df,"\n")
print("\nOne-hot encoded DataFrame for DecisionTreeClassifier and MLPClassifier:")
print(df1,"\n")
print("\nOne-hot encoded DataFrame for HistGradientBoostingClassifier:")
print(df2)
```
